```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

try:
    import wandb
    wandb_available = True
except ImportError:
    print("wandb not installed, logging to console only.")
    wandb_available = False

# DEVICE CONFIGURATION
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
if device.type == "cuda":
    print("GPU Name:", torch.cuda.get_device_name(0))
    print("Memory Allocated:", round(torch.cuda.memory_allocated(0)/1024**2,1),"MB")
    print("Memory Cached:", round(torch.cuda.memory_reserved(0)/1024**2,1),"MB")

# DATA LOADING AND PREPROCESSING
def load_fashion_data(train_ratio=0.9):
    """Load Fashion-MNIST with train/val/test split and light augmentations."""
    transform_train = transforms.Compose([
        transforms.RandomCrop(28, padding=2),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    transform_test = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    
    train_full = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform_train)
    test_set = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform_test)
    
    n_train = int(train_ratio * len(train_full))
    n_val = len(train_full) - n_train
    train_set, val_set = random_split(train_full, [n_train, n_val])
    
    return train_set, val_set, test_set

class FashionMNISTDataset(Dataset):
    """Custom dataset returning image, class label, and ink value."""
    def __init__(self, dataset):
        self.dataset = dataset
        
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        ink = img.mean()
        return img, label, ink

# MODEL DEFINITION
class MultiTaskCNN(nn.Module):
    """CNN with shared backbone and two heads (classification + regression)."""
    def __init__(self, dropout=0.3):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(1,32,3,padding=1), nn.BatchNorm2d(32), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout(dropout),
            
            nn.Conv2d(32,64,3,padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout(dropout),
            
            nn.Conv2d(64,128,3,padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2), nn.Dropout(dropout)
        )
        self.flatten = nn.Flatten()
        self.shared_fc = nn.Linear(128*3*3,128)
        self.class_head = nn.Linear(128,10)
        self.reg_head = nn.Linear(128,1)
        
    def forward(self, x, return_features=False):
        feats = []
        out = x
        for layer in self.features:
            out = layer(out)
            if isinstance(layer, nn.MaxPool2d):
                feats.append(out.clone())
        out = self.flatten(out)
        shared = F.relu(self.shared_fc(out))
        class_logits = self.class_head(shared)
        reg_out = self.reg_head(shared)
        if return_features:
            return class_logits, reg_out, feats
        return class_logits, reg_out

# LOSS AND EVALUATION
def compute_loss(class_logits, reg_out, labels, ink_targets, λ1, λ2):
    ce = F.cross_entropy(class_logits, labels)
    mse = F.mse_loss(reg_out.squeeze(), ink_targets)
    return λ1*ce + λ2*mse, ce, mse

def evaluate(model, dataloader):
    model.eval()
    total, correct = 0, 0
    mae, rmse = 0.0, 0.0
    with torch.no_grad():
        for imgs, labels, ink in dataloader:
            imgs, labels, ink = imgs.to(device), labels.to(device), ink.to(device)
            class_logits, reg_out = model(imgs)
            preds = class_logits.argmax(dim=1)
            total += labels.size(0)
            correct += (preds == labels).sum().item()
            mae += torch.abs(reg_out.squeeze() - ink).sum().item()
            rmse += torch.sum((reg_out.squeeze() - ink)**2).item()
    acc = correct / total
    mae /= total
    rmse = np.sqrt(rmse / total)
    return acc, mae, rmse

# FEATURE MAP VISUALIZATION WITH ORIGINAL IMAGE
def plot_feature_maps_with_input(model, dataloader, num_images=3):
    """
    Visualize intermediate feature maps for num_images test images
    along with the original image for comparison.
    """
    model.eval()
    imgs, _, _ = next(iter(dataloader))
    imgs = imgs[:num_images].to(device)
    
    # Convert original images to CPU and unnormalize for display
    imgs_cpu = imgs.cpu().numpy()
    imgs_disp = (imgs_cpu * 0.5) + 0.5  # reverse normalization (mean=0.5, std=0.5)
    
    with torch.no_grad():
        _, _, feats = model(imgs, return_features=True)
    
    for i in range(num_images):
        num_layers = len(feats)
        plt.figure(figsize=(3*(num_layers+1), 3))
        
        # Plot original image first
        plt.subplot(1, num_layers+1, 1)
        plt.imshow(imgs_disp[i][0], cmap='gray')
        plt.title('Original')
        plt.axis('off')
        
        # Plot feature maps (mean over channels)
        for j, fmap in enumerate(feats):
            fmap_img = fmap[i].cpu().numpy()
            fmap_mean = np.mean(fmap_img, axis=0)
            plt.subplot(1, num_layers+1, j+2)
            plt.imshow(fmap_mean, cmap='viridis')
            plt.title(f'Layer {j+1}')
            plt.axis('off')
        
        plt.suptitle(f"Feature Maps for Test Image {i+1}")
        plt.show()

# MULTI-EXPERIMENT TRAINING LOOP
# Load data
train_raw, val_raw, test_raw = load_fashion_data()
train_set = FashionMNISTDataset(train_raw)
val_set = FashionMNISTDataset(val_raw)
test_set = FashionMNISTDataset(test_raw)

# Hyperparameter combinations
experiments = [
    {"lambda_class":1,"lambda_reg":1,"optimizer":"Adam","lr":1e-3,"dropout":0.3},
    {"lambda_class":2,"lambda_reg":1,"optimizer":"Adam","lr":1e-3,"dropout":0.3},
    {"lambda_class":1,"lambda_reg":2,"optimizer":"Adam","lr":1e-3,"dropout":0.3},
    {"lambda_class":0.5,"lambda_reg":1,"optimizer":"SGD","lr":5e-4,"dropout":0.3},
    {"lambda_class":1,"lambda_reg":0.5,"optimizer":"AdamW","lr":1e-3,"dropout":0.3}
]

best_acc = 0.0
best_rmse = float('inf')
best_acc_model = None
best_rmse_model = None

for run_id, cfg in enumerate(experiments):
    print(f"\n=== Experiment {run_id+1}/{len(experiments)} ===")
    print(cfg)
    
    # Dataloaders
    train_loader = DataLoader(train_set, batch_size=64, shuffle=True)
    val_loader = DataLoader(val_set, batch_size=64)
    test_loader = DataLoader(test_set, batch_size=64)
    
    # Model
    model = MultiTaskCNN(dropout=cfg["dropout"]).to(device)
    
    # Optimizer
    if cfg["optimizer"]=="Adam":
        optimizer = torch.optim.Adam(model.parameters(), lr=cfg["lr"])
    elif cfg["optimizer"]=="SGD":
        optimizer = torch.optim.SGD(model.parameters(), lr=cfg["lr"], momentum=0.9)
    else:
        optimizer = torch.optim.AdamW(model.parameters(), lr=cfg["lr"])
    
    # Initialize wandb
    if wandb_available:
        wandb.init(project="multitask-fashionmnist", config=cfg, reinit=True)
    
    # Training
    epochs = 10
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for imgs, labels, ink in train_loader:
            imgs, labels, ink = imgs.to(device), labels.to(device), ink.to(device)
            optimizer.zero_grad()
            class_logits, reg_out = model(imgs)
            loss, ce, mse = compute_loss(class_logits, reg_out, labels, ink, cfg["lambda_class"], cfg["lambda_reg"])
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        val_acc, val_mae, val_rmse = evaluate(model, val_loader)
        
        if wandb_available:
            wandb.log({
                "train_loss": running_loss/len(train_loader),
                "val_acc": val_acc,
                "val_mae": val_mae,
                "val_rmse": val_rmse,
                "epoch": epoch+1
            })
        
        print(f"Epoch [{epoch+1}/{epochs}] Loss: {running_loss/len(train_loader):.4f} "
              f"Val Acc: {val_acc:.4f} Val MAE: {val_mae:.4f} Val RMSE: {val_rmse:.4f}")
    
    # Update best models
    if val_acc > best_acc:
        best_acc = val_acc
        best_acc_model = model
    if val_rmse < best_rmse:
        best_rmse = val_rmse
        best_rmse_model = model
    
    if wandb_available:
        wandb.finish()

# TEST EVALUATION FOR BEST MODELS
test_loader = DataLoader(test_set, batch_size=64)
acc1, mae1, rmse1 = evaluate(best_acc_model, test_loader)
acc2, mae2, rmse2 = evaluate(best_rmse_model, test_loader)

print("\nBest Validation Accuracy Model → Test Results:")
print(f"Accuracy: {acc1:.4f}, MAE: {mae1:.4f}, RMSE: {rmse1:.4f}")

print("\nBest Validation RMSE Model → Test Results:")
print(f"Accuracy: {acc2:.4f}, MAE: {mae2:.4f}, RMSE: {rmse2:.4f}")

# FEATURE MAP VISUALIZATION
plot_feature_maps_with_input(best_acc_model, test_loader)
```
