```python
# MULTI-TASK CNN ON FASHION-MNIST

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import numpy as np
import random, itertools, wandb, pandas as pd

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# 1. Custom Dataset
class FashionMNISTDataset(Dataset):
    def __init__(self, data, augment=False):
        # mean and std of FashionMNIST
        mean, std = 0.2860, 0.3530

        if augment:
            self.transform = transforms.Compose([
                transforms.RandomCrop(28, padding=2),
                transforms.RandomRotation(10),
                transforms.ToTensor(),
                transforms.Normalize((mean,), (std,))
            ])
        else:
            self.transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize((mean,), (std,))
            ])

        self.data = data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img, label = self.data[idx]
        img = self.transform(img)
        # ink = average normalized pixel intensity (mean of pixel values)
        # ink = img.mean().item()
        ink = torch.tensor(img.mean().item(), dtype=torch.float32)
        return img, label, ink

# 2. Data Loading and Splitting
def load_fashion_data():
    base_train = datasets.FashionMNIST(root="./data", train=True, download=True)
    test = datasets.FashionMNIST(root="./data", train=False, download=True)

    val_size = int(0.1 * len(base_train))
    train_size = len(base_train) - val_size
    train_raw, val_raw = random_split(base_train, [train_size, val_size])

    train = FashionMNISTDataset(train_raw, augment=True)
    val = FashionMNISTDataset(val_raw, augment=False)
    test = FashionMNISTDataset(test, augment=False)
    return train, val, test

train_data, val_data, test_data = load_fashion_data()

# 3. Model Definition
class MultiTaskCNN(nn.Module):
    def __init__(self, num_convs=3, num_filters=32, dropout=0.3):
        super().__init__()
        layers = []
        in_ch = 1
        self.feature_maps = []

        # Build convolutional backbone
        for i in range(num_convs):
            layers += [
                nn.Conv2d(in_ch, num_filters, kernel_size=3, padding=1),
                nn.BatchNorm2d(num_filters),
                nn.ReLU(),
                nn.MaxPool2d(2),
                nn.Dropout(dropout),
            ]
            in_ch = num_filters
            num_filters *= 2

        self.conv = nn.Sequential(*layers)
        self.flatten = nn.Flatten()

        # Dynamically infer flatten size (safe version)
        try:
            with torch.no_grad():
                dummy = torch.zeros(1, 1, 28, 28)
                dummy_out = self.conv(dummy)
                flatten_dim = dummy_out.view(1, -1).shape[1]
        except Exception as e:
            print("Warning: dummy forward failed during init, using default 7x7 size.")
            print("Error:", e)
            flatten_dim = in_ch * 7 * 7

        # Shared + separate heads
        self.fc_shared = nn.Linear(flatten_dim, 128)
        self.class_head = nn.Linear(128, 10)
        self.reg_head = nn.Linear(128, 1)

    def forward(self, x):
        self.feature_maps = []
        for layer in self.conv:
            x = layer(x)
            if isinstance(layer, nn.ReLU):  # capture after each block
                self.feature_maps.append(x)
        flat = self.flatten(x)
        shared = F.relu(self.fc_shared(flat))
        class_logits = self.class_head(shared)
        reg_out = self.reg_head(shared)
        return class_logits, reg_out, self.feature_maps

# 4. Feature Map Visualization
def plot_feature_maps(model, img, title):
    model.eval()
    with torch.no_grad():
        _, _, feature_maps = model(img.unsqueeze(0).to(device))
    fig, axes = plt.subplots(1, len(feature_maps)+1, figsize=(16, 4))
    axes[0].imshow(img.squeeze(), cmap='gray')
    axes[0].set_title("Input")
    for i, fmap in enumerate(feature_maps):
        axes[i+1].imshow(fmap[0, 0].cpu(), cmap='viridis')
        axes[i+1].set_title(f"Block {i+1}")
    fig.suptitle(title)
    plt.tight_layout()
    plt.show()

# 5. Training + Validation
def train_and_validate(config, run_id):
    print(f"\n=== Experiment {run_id+1}/10 ===\n{config}")
    wandb.init(project="multi_task_fashion", config=config, reinit=True)

    model = MultiTaskCNN(config["num_convs"], config["num_filters"], config["dropout"]).to(device)
    opt = getattr(torch.optim, config["optimizer"])(model.parameters(), lr=config["lr"])
    ce_loss = nn.CrossEntropyLoss()
    mse_loss = nn.MSELoss()

    train_loader = DataLoader(train_data, batch_size=config["batch_size"], shuffle=True)
    val_loader = DataLoader(val_data, batch_size=config["batch_size"])

    best_acc, best_rmse = 0, float("inf")
    best_model_acc, best_model_rmse = None, None

    for epoch in range(8):
        model.train()
        total_loss, total_ce, total_mse = 0, 0, 0

        for imgs, labels, inks in train_loader:
            imgs, labels, inks = imgs.to(device), labels.to(device), inks.unsqueeze(1).to(device)
            opt.zero_grad()
            logits, reg_out, _ = model(imgs)
            ce = ce_loss(logits, labels)
            mse = mse_loss(reg_out, inks)
            loss = config["lambda_class"] * ce + config["lambda_reg"] * mse
            loss.backward()
            opt.step()
            total_loss += loss.item()
            total_ce += ce.item()
            total_mse += mse.item()

        # Validation
        model.eval()
        correct, total, val_mse = 0, 0, 0
        with torch.no_grad():
            for imgs, labels, inks in val_loader:
                imgs, labels, inks = imgs.to(device), labels.to(device), inks.unsqueeze(1).to(device)
                logits, reg_out, _ = model(imgs)
                preds = logits.argmax(1)
                correct += (preds == labels).sum().item()
                total += labels.size(0)
                val_mse += mse_loss(reg_out, inks).item()

        acc = correct / total
        rmse = np.sqrt(val_mse / len(val_loader))

        wandb.log({
            "epoch": epoch,
            "train_loss": total_loss/len(train_loader),
            "train_CE": total_ce/len(train_loader),
            "train_MSE": total_mse/len(train_loader),
            "val_acc": acc,
            "val_rmse": rmse
        })

        print(f"Epoch {epoch+1}/8 | Loss={total_loss/len(train_loader):.4f} | "
              f"Acc={acc:.4f} | RMSE={rmse:.4f}")

        if acc > best_acc:
            best_acc, best_model_acc = acc, model.state_dict()
        if rmse < best_rmse:
            best_rmse, best_model_rmse = rmse, model.state_dict()

    wandb.finish()
    return {"config": config, "best_acc": best_acc, "best_rmse": best_rmse,
            "model_acc": best_model_acc, "model_rmse": best_model_rmse}

# 6. Hyperparameter Sweep
param_grid = {
    "lr": [1e-3, 5e-4],
    "dropout": [0.2, 0.4],
    "num_convs": [3],
    "num_filters": [16, 32],
    "batch_size": [64, 128],
    "optimizer": ["Adam", "AdamW"],
    "lambda_class": [0.5, 1, 2],
    "lambda_reg": [0.5, 1, 2],
}

param_combinations = random.sample(list(itertools.product(*param_grid.values())), 10)

results = []
best_acc, best_rmse = 0, float("inf")
best_model_acc, best_model_rmse = None, None

for i, combo in enumerate(param_combinations):
    config = dict(zip(param_grid.keys(), combo))
    result = train_and_validate(config, i)
    results.append(result)

    if result["best_acc"] > best_acc:
        best_acc, best_model_acc = result["best_acc"], result["model_acc"]
        best_acc_cfg = result["config"]
    if result["best_rmse"] < best_rmse:
        best_rmse, best_model_rmse = result["best_rmse"], result["model_rmse"]
        best_rmse_cfg = result["config"]

# 7. Test Evaluation
def evaluate(model_state, tag, cfg):
    model = MultiTaskCNN(
        num_convs=cfg["num_convs"],
        num_filters=cfg["num_filters"],
        dropout=cfg["dropout"]
    ).to(device)
    model.load_state_dict(model_state)
    model.eval()

    mse_loss = nn.MSELoss()
    correct, total, mse = 0, 0, 0
    test_loader = DataLoader(test_data, batch_size=128)

    with torch.no_grad():
        for imgs, labels, inks in test_loader:
            imgs, labels, inks = imgs.to(device), labels.to(device), inks.unsqueeze(1).to(device)
            logits, reg_out, _ = model(imgs)
            preds = logits.argmax(1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)
            mse += mse_loss(reg_out, inks).item()

    print(f"[{tag}] Test Accuracy: {correct/total:.4f}, Test RMSE: {np.sqrt(mse/len(test_loader)):.4f}")

evaluate(best_model_acc, "Best Accuracy Model", best_acc_cfg)
evaluate(best_model_rmse, "Best RMSE Model", best_rmse_cfg)

# 8. Summary Table
summary = []
for r in results:
    row = r["config"].copy()
    row["Val_Acc"] = r["best_acc"]
    row["Val_RMSE"] = r["best_rmse"]
    summary.append(row)
df = pd.DataFrame(summary)
print("\n===== EXPERIMENT SUMMARY =====")
print(df.sort_values(by="Val_Acc", ascending=False).reset_index(drop=True))

# 9. Feature Map Visualization (3 test images)
model_state = None
cfg = None

if 'best_model_acc' in globals() and best_model_acc is not None and 'best_acc_cfg' in globals():
    model_state = best_model_acc
    cfg = best_acc_cfg
elif 'best_model_rmse' in globals() and best_model_rmse is not None and 'best_rmse_cfg' in globals():
    model_state = best_model_rmse
    cfg = best_rmse_cfg

if cfg is not None and model_state is not None:
    model = MultiTaskCNN(num_convs=cfg["num_convs"],
                         num_filters=cfg["num_filters"],
                         dropout=cfg["dropout"]).to(device)
    try:
        model.load_state_dict(model_state)
    except RuntimeError as e:
        # If exact loading fails (e.g., slight mismatch), try non-strict load and warn.
        print("Warning: strict state_dict load failed:", e)
        print("Retrying load_state_dict with strict=False")
        model.load_state_dict(model_state, strict=False)
else:
    # Fall back to default model if no saved state available
    print("No matching saved model config found — using default MultiTaskCNN instance.")
    model = MultiTaskCNN().to(device)
    if 'best_model_acc' in globals() and best_model_acc is not None:
        try:
            model.load_state_dict(best_model_acc, strict=False)
        except Exception as e:
            print("Could not load checkpoint into default model:", e)

for i in range(3):
    img, _, _ = test_data[i]
    plot_feature_maps(model, img, f"Test Image {i+1}")
```
